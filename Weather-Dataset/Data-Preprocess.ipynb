# Weather Data Preprocess

```python
import pandas as pd
```

```python
# Read daily data
daily_data = pd.read_csv('open-meteo-daily-data-indramayu.csv')
# Read hourly data
hourly_data = pd.read_csv('open-meteo-hourly-data-indramayu.csv')
```

```python
daily_data['time'] = pd.to_datetime(daily_data['time'])
hourly_data['time'] = pd.to_datetime(hourly_data['time'])
```

```python
daily_data['date'] = daily_data['time'].dt.date
hourly_data['date'] = hourly_data['time'].dt.date
```

```python
daily_data.rename(columns={'weather_code (wmo code)': 'daily_weather_code (wmo code)'}, inplace=True)
```

```python
daily_data.head()
```

```python
hourly_data.head()
```

```python
merged_data = hourly_data.merge(daily_data, on='date')
```

```python
merged_data.head()
```

```python
merged_data.drop(columns=['date', 'time_y'], inplace=True)
```

```python
merged_data.head()
```

```python
merged_data.to_csv('merged_weather_data.csv', index=False)

print("Merging completed. Data saved to 'merged_weather_data.csv'")
```

# Merge with marine

```python
weather_data = merged_data

# Read preprocessed marine data
marine_data = pd.read_csv('merged_marine_data.csv')
```

```python
weather_data.head()
```

```python
marine_data['time_x'] = pd.to_datetime(marine_data['time_x'])
```

```python
marine_data.head()
```

```python
merged_data = weather_data.merge(marine_data, on='time_x')
```

```python
merged_data.head()
```

```python
merged_data.rename(columns={'time_x': 'time'}, inplace=True)
```

```python
merged_data.to_csv('merged_weather_marine_data_1jan-29feb.csv', index=False)

print("Merging completed. Data saved to 'merged_weather_data.csv'")
```
