```python
import pandas as pd
```

```python
# Read daily data
daily_data = pd.read_csv('open-meteo-daily-data-indramayu-marine.csv')
# Read hourly data
hourly_data = pd.read_csv('open-meteo-hourly-data-indramayu-marine.csv')
```

```python
daily_data['time'] = pd.to_datetime(daily_data['time'])
hourly_data['time'] = pd.to_datetime(hourly_data['time'])
```

```python
daily_data['date'] = daily_data['time'].dt.date
hourly_data['date'] = hourly_data['time'].dt.date
```

```python
daily_data.head()
```

```python
hourly_data.head()
```

```python
merged_data = hourly_data.merge(daily_data, on='date')
```

```python
merged_data.head()
```

```python
merged_data.drop(columns=['date', 'time_y'], inplace=True)
```

```python
merged_data.head()
```

```python
columns_order = [
    'time_x', 'wave_height (m)', 'wave_height_max (m)', 'wave_direction (°)', 'wave_direction_dominant (°)', 'wave_period (s)',
    'wave_period_max (s)',
    'wind_wave_height (m)', 'wind_wave_height_max (m)', 'wind_wave_direction (°)', 'wind_wave_direction_dominant (°)', 'swell_wave_height (m)',
    'swell_wave_height_max (m)',
    'swell_wave_direction (°)', 'swell_wave_direction_dominant (°)'
]
```

```python
merged_data = merged_data[columns_order]
```

```python
merged_data.head()
```

```python
merged_data.to_csv('merged_marine_data.csv', index=False)

print("Merging completed. Data saved to 'merged_marine_data.csv'")
```
