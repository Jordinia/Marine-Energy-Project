```python
import pandas as pd
```

```python
# Read daily data
daily_data = pd.read_csv('open-meteo-daily-data-indramayu.csv')
# Read hourly data
hourly_data = pd.read_csv('open-meteo-hourly-data-indramayu.csv')
```

```python
daily_data['time'] = pd.to_datetime(daily_data['time'])
hourly_data['time'] = pd.to_datetime(hourly_data['time'])
```

```python
daily_data['date'] = daily_data['time'].dt.date
hourly_data['date'] = hourly_data['time'].dt.date
```

```python
daily_data.rename(columns={'weather_code (wmo code)': 'daily_weather_code (wmo code)'}, inplace=True)
```

```python
daily_data.head()
```

```python
hourly_data.head()
```

```python
merged_data = hourly_data.merge(daily_data, on='date')
```

```python
merged_data.head()
```

```python
merged_data.drop(columns=['date'], inplace=True)
```

```python
merged_data.to_csv('merged_weather_data.csv', index=False)

print("Merging completed. Data saved to 'merged_weather_data.csv'")
```
